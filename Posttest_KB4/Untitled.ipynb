# Menentukan Feature dan Target 

```python
# Pilih kolom numerik sebagai fitur (misalnya 'Age' dan 'GPA')
X = df[['time_left', 'ct_score']]
# Kolom label kategorikal
y = df['round_winner']
```

# dataset dibagi menjadi training set dan testing set dengan proporsi 70:30

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# lakukan normalisasi data pada salah satu attribute menggunakan Min Max

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# copy dataset

```python
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

# minimax scaller

```python
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_rep)
```
